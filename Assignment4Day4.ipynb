```python
def DOZscore(sample):
    import numpy as np
    outliers = []
    thres = 3
    mean = np.mean(sample)
    std = np.std(sample)
    for i in sample:
        z_score=(i-mean)/std
        if(np.abs(z_score)>thres):
            outliers.append(i)

    return outliers

Data_TotalVolume = df['Total Volume']
outliers_TV = DOZscore(Data_TotalVolume)
print(outliers_TV)

'''Replacing the outliers with their medians because outliers
affect the central tendencies of the data and hence handling
the outliers
'''
```

```python
median = np.median(Data_TotalVolume)
for i in outliers_TV:
    c = np.where(Data_TotalVolume==i,median,Data_TotalVolume)
    print("Data:",Data_TotalVolume)
    print("New data:",c)
     
```

```python
type_totals = df.groupby('type')['Total Bags'].sum()
```

```python
type_totals
'''.groupby('ProductType'): This is the grouping operation. It tells pandas to group the data in the DataFrame based on the unique values in the 'ProductType' column. Each unique value becomes a group.

['TotalSold']: This specifies the column on which the aggregation will be performed within each group. In this case, you want to sum the 'TotalSold' values within each product type group.

.sum(): This is the aggregation function. It calculates the sum of the 'TotalSold' values for each group (product type).'''
```

```python
plt.pie(type_totals,labels=type_totals.index,colors=('blue','pink'))
```
