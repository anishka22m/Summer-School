```python
import random as r
```

```python
prod = {'p1': 10, 'p2': 15, 'p3': 20, 'p4': 25, 'p5': 30, 'p6': 35, 'p7': 50}
```

```python
def sizerange(values):
    return r.randint(2, len(values))
```

```python
def pricerange(LL, UL):
    while True:
        price = r.randint(LL, UL)
        if LL <= price <= UL:
            return price
```

```python
def ComboFinder(products, totalprice):
    values = list(products.values())
    allcombos = []
    for i in range(1000):  # Increase the number of attempts
        combo = r.choices(values, k=sizerange(values))
        combo_sum = sum(combo)
        if combo_sum == totalprice:
            combo_keys = [key for key, value in products.items() if value in combo]
            allcombos.append(combo_keys)
    return allcombos
```

```python
all_combos = []
for i in range(10):
    all_combos.extend(ComboFinder(prod, pricerange(90, 110)))
```

Convert inner lists to sets and take the union

```python
union_of_combos = set(map(frozenset, all_combos))
```

```python
print("Union of all distinct combos:")
for acombo in union_of_combos:
    print(list(acombo))
print(len(union_of_combos))
```
